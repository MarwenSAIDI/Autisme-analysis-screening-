```python
def clean_round1(text):
    text = str(text).lower()
    text = re.sub("s/he","she or he", text)
    text = re.sub("/"," or ", text)
    text = re.sub("e.g.","for example", text)
    text = re.sub("[?“”-]","", text)
    return text
round_1 = lambda x:clean_round1(x)
```

```python
cast.Questions = cast.Questions.apply(round_1)
```

```python
asq.Questions = asq.Questions.apply(round_1)
```
